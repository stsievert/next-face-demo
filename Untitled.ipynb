```python
import face_recognition
import numpy as np
import cv2
import face_api_local
import plot_util
from matplotlib.image import imread
```

```python
img = "test-faces/internet/3.png"
img = cv2.imread(img)
rgb, rgba = plot_util.process_image(img, 500)
img_flipped = np.flipud(rgb)
```

```python
face_locations = face_recognition.face_locations(rgb)
```

```python
face_locations
```

```python
face_landmarks_list = face_recognition.face_landmarks(rgb)[0]
```

```python
face_landmarks_list
```

```python
face_points = []
for key, values in face_landmarks_list.items():
    for cord in values:
        newPoint = [cord[0], cord[1]]
        face_points.append(newPoint)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(*zip(*face_points), marker='o', color='b', ls='')
plt.imshow(rgb, interpolation='nearest')
```
